5. Use your ChatBot session to help understand the difference between the following and then provide your own paraphrasing summarization of that difference

an "attribute", such as df.shape which does not end with ()

and a "method", such as df.describe() which does end with ()

For an attribute, it stores a value / property of an object. For example, df.shape stores the (no. of rows, no. of columns) as its value. It does not end with (), implies that it does not require any input to work -- thus there is no calculation or computation involved. It can simply be accessed like any other python variables.

For a method, it is a function executed along together with the object. It does not store a value, but instead RETURNS values / EXECUTE instructions. For instance, df.describe() executes a summarization of the dataset and returns the 'description' as the output. Note that there is nothing inside the '()'. This is because it is optional to pass values in the paraentheses in some cases. E.g. df.describe(include='all') and df.describe both works, but the former one performed a customized instruction instead of the latter default one.
