# Research Question 1
**Are older people less dissatisfied with their appearance?**


## Variables to be Explored

### Outcome Variable (Main Variable to Investigate)

- **Unattractiveness (self-perceived)**: `PSYCH_body_self_image_questionnaire_think_unattractive`
    - Relates to how a person views their unattractiveness.
    - **Visualization method**: Scatter diagram - provides an overview of self-perceived unattractiveness across age

### Additional Side Variables (Optional Explorations)

These variables relate to self-perceived body image and opinions about one's body size:

- **PSYCH_body_self_image_questionnaire_well_proportioned**: *“Having a well-proportioned body is important to me.”*
- **PSYCH_body_self_image_questionnaire_looks_fat**: *“I think my body looks fat in clothes.”*
- **PSYCH_body_self_image_questionnaire_pay_attention**: *“I pay careful attention to my face and hair, so that I will look good.”*
- **PSYCH_body_self_image_questionnaire_different_height**: *“I wish I were a different height.”*
- **PSYCH_body_self_image_questionnaire_depressed**: *“I feel depressed about my body.”*
- **PSYCH_body_self_image_questionnaire_thinner**: *“I wish I were thinner.”*
- **PSYCH_body_self_image_questionnaire_overweight**: *“My body is overweight.”*

  - **Visualization method**: Scatter diagram - illustrates people’s self-evaluations of body size and appearance with different ages.

---

## Predictor Variable

- **Age**: `DEMO_age`
    - Represents participant age.
    - **Rationale**: There is a common belief that younger individuals are more concerned about appearance, while older adults may feel "less attractive" relative to contemporary beauty standards.
    - **Visualization method**: Histogram - useful for assessing the age distribution of participants.

---

## Analysis Plan

1. **Mean Difference Comparison**: Investigate mean differences in body image scores across age to identify trends in body dissatisfaction.
2. **Linear Regression**: Use linear regression to assess the relationship between age and body image scores and obtain the correlation (R-value) to see the strength of association.
3. **Visualizations**:
   - **Box Plots**: To compare body dissatisfaction scores by age groups (e.g. Age 0-9, 10-19...)
   - **Histograms**: To show the age distribution.

---

## Assumptions

1. **General Assumptions**:
   - The predictor variable (age) is measured without error.

2. **Linear Regression Assumptions**:
   - The errors $\epsilon_i$ are normally distributed.
   - The errors $\epsilon_i$ are homoscedastic (variance does not change with age).
   - The relationship is linear, so the linear regression model can approximate the trend.
   - Errors are statistically independent.
   - Errors are unbiased relative to the expected outcome.

---

## Hypotheses

1. **Hypothesis 1**: Age has no effect on people's level of satisfaction with their appearance.

---

## Discussion and Preliminary Analysis


1. **Hypothesis 1: Age and Appearance Dissatisfaction**
   - **Analysis Approach**: Conduct a linear regression to examine the relationship between age and self-perceived unattractiveness. A positive regression coefficient would indicate that self-perceived unattractiveness increases with age, while a negative coefficient would suggest it decreases.
   - **Relevance**: Validating this hypothesis would help highlight if body dissatisfaction is more prevalent in certain age demographicsl, like youngsters.
   - **Way of testing**: Confidence interval: calculate the confidence interval of the slope (confidence interval =  Slope x 1.96 x Standard Error of the Slope). Then, we can check whether the slope 0 is in the range of the confidence interval. If not, we can reject the null hypothesis that age has no effect on people's level of satisfaction with their appearance.

### Additional Side Variables (Optional Explorations)

## Predictor Variables

    - Represents participant gender.
    - **Rationale**: It’s often suggested that males may care less about appearance than females; testing this hypothesis can help explore if this is truly the case or a stereotype.
    - **Visualization method**: Pie chart - displays gender distribution across the dataset.

---

## Analysis Plan

1. **Mean Difference Comparison**: Investigate mean differences in body image scores across gender to identify trends in body dissatisfaction.
2. **Visualizations**:
   - **Pie Charts**: To visualize gender distribution of the participants.

---

## Assumptions

1. **General Assumptions**:
   - The predictor variable (gender) is measured without error (i.e. no one accidentally chose the wrong gender)
   
---

## Hypotheses

1. **Hypothesis 2**: Gender has no effect on people's level of satisfaction with their appearance.

---

## Discussion and Preliminary Analysis

1. **Hypothesis 2: Gender and Appearance Dissatisfaction**
   - **Analysis Approach**: Using p-value hypothesis testing.
   - **Relevance**: Findings would contribute to understanding whether body dissatisfaction issues are gender-specific.
   - **Way of testing**: p-value hypothesis testing: first, find the actual observed difference in mean satisfaction scores between male and female. Simulate 10000 cases in accordance with the null hypothesis. It can be achieved by randomly shuffling the label of male and female for each simulation. For each simulation, we find their difference in mean. Then, we use 10000 of those means to create a simulated difference-in-mean distribution. After that, we calculate our p-value, which is defined as the proportion of the simulated difference-in-mean distribution that is as or more extreme than the observed test statistic (the actual observed difference in mean satisfaction score) under the null hypothesis.
