Q1 Ans: 
- Predictor: it is a variable usually denoted as x and is considered the independent variable. It serves as the 'input' to the straight line function from the linear regression.
- Outcome: it is a variable usually denoted as y and is considered the dependent variable. It is the 'output' or 'guess' from the straight line function from the linear regression.
- Slope: it is the amount of units that the value of outcome y would change per unit change of x. It essentially is defined as the rate of change for the linear relationship.
- Error term: it is the variability of the model to the true value of y, capturing the influence of other factors affecting the outcome. It is a random variable drawn from the normal distribution. The normal distribution of the error term has a mean of zero which implies the linear regression model is unbiased.
---
The below plots illustrates the nature of linear regression model:

In the regression model below, we see that the scatter diagram shows the data points in the dataset. It is NOT a straightline, implying that a model fitting in the graph would not capture all the data points. Thus, it explains the need of an error term for the variability of our model.

As for the second graph, we can see that the error term distribution is normally distributed. 


````python
import numpy as np
import plotly.graph_objects as go

# Set random seed for reproducibility
np.random.seed(1)

# Parameters for the linear regression
beta0 = 5  # intercept
beta1 = 2  # slope
n = 100 # number of samples

# Generate predictor variable (X)
X = np.random.rand(n) * 10  # X values from 0 to 10

# Generate normally distributed error term (ε)
epsilon = np.random.normal(loc=0, scale=1, size=n)  # mean=0, std=1

# Generate outcome variable (Y)
Y = beta0 + beta1 * X + epsilon  # Linear relation with noise

# Create a figure with plotly.graph_objects
fig = go.Figure()

# Scatter plot for data points
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Data',
                         marker=dict(color='black')))

# Add the regression line to the plot
x_range = np.linspace(0, 10, 100)
y_range = beta0 + beta1 * x_range
fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression Line',
                         line=dict(color='red')))

# Set layout for the plot
fig.update_layout(
    title='Simple Linear Regression Model',
    xaxis_title='Predictor Variable (X)',
    yaxis_title='Outcome Variable (Y)'
)

# Show the plot
fig.show(renderer="png")
````

````python
import statsmodels.formula.api as smf  # This library allows us to fit statistical model. 
import plotly.express as px  # This is the plotly library we will be using.
import pandas as pd


# Combining simulated x and y into a pandas dataframe object
df = pd.DataFrame({'X': X, 'Y': Y})

#These two lines prepare 
model_data_specification = smf.ols("Y~X", data=df) # This line creates an Ordinary Least Squares model with y being the outcome and x being the predictor, with x and y being columns of the dataframe.
fitted_model = model_data_specification.fit() # This step fits the above created model to the data with the best-fit line. 

# The below provides information and values related to the fitted model.
fitted_model.summary()  # Generates a detailed summary report of the regression results.
fitted_model.summary().tables[1]  # Provides only the estimated coefficients for the intercept / predictor variable
fitted_model.params  # Provides the estimated coefficients of the model (intercept and slope)
fitted_model.params.values  # Extract the values as a NumPy array without the labels.
fitted_model.rsquared  # Returns the r-squared value. A 0.7 r-squared value means 70% of the variation of the outcomes can be explained by the model.

# This processes the dataframe using plotly to create an object of scatter plot, including the trendline "best fit line" using OLS.
df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='X',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. X') # The OLS trendline generated by plotly

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['X'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'") # Adding the OLS trendline generated from statsmodel


fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

The line chosen is the best-fitting line which minimizes the sum of the squared differences between the observed values and the predicted values. 

By squaring the residuals of the estimated model, we give more weight to those larger discrepancies such that the model captures more outlier accurately.

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n, x_min, x_range, beta0, beta1, sigma = 20, 5, 5, 2, 3, 5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

# Use consistent column names
df = pd.DataFrame({'x': x, 'y': Y})  # Changed 'Y' to 'y' for consistency
model_data_specification = smf.ols("y ~ x", data=df)  # Changed to use 'y'
fitted_model = model_data_specification.fit()

df['Data'] = 'Data'  # hack to add data to legend 
fig = px.scatter(df, x='x', y='y', color='Data',  # Changed 'Y' to 'y'
                 trendline='ols', title='Y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0) + ' + ' + str(beta1) + ' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],  # Changed 'Y' to 'y'
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)

# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()] * 2, mode='lines',  # Changed 'Y' to 'y'
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer="png")  # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

### 6. Explain why the first expression below can be interpreted as "the proportion of variation in (outcome) Y explained by the model (i.e. _fitted_model.fittedvalues_)"; and therefore, why _fitted_model.rsquared_ can be interpreted as a measure of the accuracy of the model; and, therefore what the two _np.corrcoef(...)[0,1]\*\*2_ expressions capture in the context of _Simple Linear Regression models_.

4. `np.corrcoef(Y,x)[0,1]**2`<br><br>

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>
> _$R^2 = 1 - \frac{\sum_{i=1}^n(Y_i-\hat y)^2}{\sum_{i=1}^n(Y_i-\bar Y)^2}$_
>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._
>

The first expression is essentially the formula for R-squared. By definition, R-squared tells us th proportion of the total variation in Y explain by the model. 

The fitted-model.rsquared thus gives us how well the independent variable explains the variation in Y. The better the independent variable explains the variation in Y, the more accurate the model we are talking about.

For the third expression, we are trying to compute the squared correlation of the actual values of Y (theoretical) and the fitted values of Y hat from the generated samples. np.corrcoef() calculates the correlation R, and by squaring it at the end we obtain R^2.

````python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import plotly.express as px
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
beta_0 = 1.65
beta_1 = 0  # No linear association
sigma = 0.37
n_samples = 1000  # Number of bootstrap samples
sample_size = 160  # Sample size for each simulation

# Load the Old Faithful Geyser dataset and get waiting times
old_faithful = sns.load_dataset('geyser')
waiting_times = old_faithful['waiting'].values[:sample_size]  # Get first 160 waiting times

# Initialize a list to collect slope coefficients
slope_coefficients = []

# Simulate samples
for _ in range(n_samples):
    # Simulate Y values under the null hypothesis
    epsilon = np.random.normal(0, sigma, size=sample_size)  # Errors
    Y = beta_0 + beta_1 * waiting_times + epsilon  # Simulated outcomes

    # Fit the Simple Linear Regression model
    X = sm.add_constant(waiting_times)  # Add intercept
    model = sm.OLS(Y, X).fit()
    
    # Collect the slope coefficient
    slope_coefficients.append(model.params[1])  # Index 1 corresponds to slope coefficient

# Convert to a DataFrame for easier handling with Plotly
slope_coefficients_df = pd.DataFrame(slope_coefficients, columns=['slope_coefficient'])

# Visualize the bootstrapped sampling distribution of the slope coefficients using Plotly
fig = px.histogram(slope_coefficients_df, x='slope_coefficient', nbins=30,
                   title='Bootstrapped Sampling Distribution of the Slope Coefficients under Null Hypothesis',
                   labels={'slope_coefficient': 'Slope Coefficient (duration ~ waiting)'})

# Add mean and confidence intervals
mean_slope = np.mean(slope_coefficients)
lower_bound = np.percentile(slope_coefficients, 2.5)
upper_bound = np.percentile(slope_coefficients, 97.5)

# Add vertical lines for mean and confidence intervals
fig.add_vline(x=mean_slope, line_color='red', line_dash='dash', 
               annotation_text='Mean', annotation_position='top right')
fig.add_vline(x=lower_bound, line_color='blue', line_dash='dash', 
               annotation_text='2.5% Percentile', annotation_position='top right')
fig.add_vline(x=upper_bound, line_color='blue', line_dash='dash', 
               annotation_text='97.5% Percentile', annotation_position='top right')

# Show the figure
fig.update_layout(xaxis_title='Slope Coefficient (duration ~ waiting)', 
                  yaxis_title='Frequency')
fig.show(renderer="png")
````

Then, we set the parameters for bootstrap as usual, and we do the simulation of sampling distribution under null -- that is, there is no difference in duration between the short and long waiting time. Since if the null hypothesis is true, then label shuffing does not matter as there is no difference between the two populations. We build a distribution of mean differences under this. Then, we calculate the p-value by checking the proportion of the simulated data that are as or more extreme than the observed mean difference.

For double bootstrapping, we draw samples with replacement for the two columns at the same time, then we calculate the mean difference between the corresponding bootstrap samples for short and long wait times. At last, we construct confidence interval using np.quantile(). 


(b) 
- Similarites: all of these models aim to provide insight in means between the two groups. 
- Differences: Permutation and double bootstrap approaches are non-parametric, while indicator variable model is.

> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
