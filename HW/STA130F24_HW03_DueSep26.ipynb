````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Create the base histogram using facets for each species
fig = px.histogram(penguins, 
                   x="flipper_length_mm", 
                   facet_col="species", 
                   color="species",
                   opacity=0.7,
                   title="Flipper Length Distribution by Species")

# Statistical overlays (mean, median, range, IQR, ±2 standard deviations)
for species in penguins['species'].unique():
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()
    
    mean = species_data.mean()
    median = species_data.median()
    std_dev = species_data.std()
    q1, q3 = species_data.quantile([0.25, 0.75])
    min_val, max_val = species_data.min(), species_data.max()
    
    # Identify the column for the species in the facet grid
    col = list(penguins['species'].unique()).index(species) + 1
    
    # Add the mean (green) and median (blue) lines
    fig.add_vline(x=mean, line_dash="dash", line_color="green", row=1, col=col)
    fig.add_vline(x=median, line_dash="solid", line_color="blue", row=1, col=col)
    
    # Add range rectangle (min to max, light purple)
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="rgba(153, 51, 255, 0.2)", line_width=0, row=1, col=col)
    
    # Add interquartile range rectangle (Q1 to Q3, light orange)
    fig.add_vrect(x0=q1, x1=q3, fillcolor="rgba(255, 165, 0, 0.4)", line_width=0, row=1, col=col)
    
    # Add ±2 std deviation rectangle (±2 SD, light red)
    fig.add_vrect(x0=mean - 2 * std_dev, x1=mean + 2 * std_dev, fillcolor="rgba(255, 0, 0, 0.2)", line_width=0, row=1, col=col)

# Add dummy traces for the legend
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         line=dict(color="green", dash="dash"), 
                         name="Mean"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         line=dict(color="blue"), 
                         name="Median"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(153, 51, 255, 0.2)", 
                         line=dict(color="purple"), 
                         name="Range"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(255, 165, 0, 0.4)", 
                         line=dict(color="orange"), 
                         name="IQR"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(255, 0, 0, 0.2)", 
                         line=dict(color="red"), 
                         name="±2 Std Dev"))

# Show the plot with the legend
fig.show(renderer="png")
````

````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import seaborn as sns

# Load the dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Create the base histogram using facets for each species
fig = px.histogram(penguins, 
                   x="flipper_length_mm", 
                   facet_col="species", 
                   color="species",
                   opacity=0.5,
                   histnorm='probability density',  # Normalize histogram
                   title="Flipper Length Distribution and KDE by Species")

# Statistical overlays (mean, median, range, IQR, ±2 standard deviations)
for species in penguins['species'].unique():
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()
    
    mean = species_data.mean()
    median = species_data.median()
    std_dev = species_data.std()
    q1, q3 = species_data.quantile([0.25, 0.75])
    min_val, max_val = species_data.min(), species_data.max()
    
    # Identify the column for the species in the facet grid
    col = list(penguins['species'].unique()).index(species) + 1
    
    # Add the mean (green) and median (blue) lines
    fig.add_vline(x=mean, line_dash="dash", line_color="green", row=1, col=col)
    fig.add_vline(x=median, line_dash="solid", line_color="blue", row=1, col=col)
    
    # Add range rectangle (min to max, light purple)
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="rgba(153, 51, 255, 0.2)", line_width=0, row=1, col=col)
    
    # Add interquartile range rectangle (Q1 to Q3, light orange)
    fig.add_vrect(x0=q1, x1=q3, fillcolor="rgba(255, 165, 0, 0.4)", line_width=0, row=1, col=col)
    
    # Add ±2 std deviation rectangle (±2 SD, light red)
    fig.add_vrect(x0=mean - 2 * std_dev, x1=mean + 2 * std_dev, fillcolor="rgba(255, 0, 0, 0.2)", line_width=0, row=1, col=col)

    # Calculate KDE using seaborn
    x_kde = np.linspace(min_val, max_val, 100)
    kde_values = sns.kdeplot(species_data, bw_adjust=0.5, common_norm=False).get_lines()[col-1].get_data()
    
    # Extract KDE x and y values
    x_kde = kde_values[0]
    y_kde = kde_values[1]

    # Normalize the y values to make sure the area under the KDE curve is equal to 1
    y_kde /= np.trapz(y_kde, x_kde)  # Normalize the area under the curve

    # Overlay KDE plot on the histogram
    fig.add_trace(go.Scatter(x=x_kde, y=y_kde, mode='lines', 
                             line=dict(color="black", width=2), 
                             name=f"KDE {species}"),
                  row=1, col=col)

# Add dummy traces for the legend
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         line=dict(color="green", dash="dash"), 
                         name="Mean"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         line=dict(color="blue"), 
                         name="Median"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(153, 51, 255, 0.2)", 
                         line=dict(color="purple"), 
                         name="Range"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(255, 165, 0, 0.4)", 
                         line=dict(color="orange"), 
                         name="IQR"))
fig.add_trace(go.Scatter(x=[None], y=[None], mode='lines', 
                         fill='toself', 
                         fillcolor="rgba(255, 0, 0, 0.2)", 
                         line=dict(color="red"), 
                         name="±2 Std Dev"))

# Show the plot with the legend
fig.show(renderer="png")
````

### Comparison of Box Plot, Histogram, and Kernel Density Estimator (KDE):

1. **Box Plot**:
   - Displays range, median, interquartile range (IQR), and outliers.
   - Useful for summarizing key statistics but does not show the actual distribution of data points.
   
2. **Histogram**:
   - Divides data into bins and shows frequency on the y-axis.
   - Provides a clear "mountain" shape to understand the distribution but lacks summary figures like the median or IQR.
   - Bin size significantly affects the visualization, and careful selection is necessary.

3. **Kernel Density Estimator (KDE)**:
   - Generates a smooth graph to represent the probability density of the data.
   - Shows the continuous shape of the distribution and is visually clearer than a histogram, but it uses density on the y-axis instead of frequency.
   - Sensitive to bandwidth selection, which affects the smoothness of the curve.

### Pros and Cons Summary:

| **Method**               | **Pros**                                                          | **Cons**                                                             |
|--------------------------|-------------------------------------------------------------------|----------------------------------------------------------------------|
| **Box Plot**              | - Highlights key statistics (median, quartiles, outliers) <br> - Compact and simple for comparison | - Does not show frequency or the actual distribution of data |
| **Histogram**             | - Shows raw frequency and distribution shape clearly <br> - Good for discrete data | - Bin size impacts accuracy <br> - No summary statistics |
| **Kernel Density Estimator (KDE)** | - Smooth, continuous representation of the data distribution <br> - No binning issues | - Requires bandwidth tuning <br> - Shows density, not raw counts |

Each method has its unique benefits: box plots are best for summary statistics and outliers, histograms provide frequency-based insights, and KDEs offer a smooth representation of the distribution.

````python
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

n = 1500
data1 = stats.uniform.rvs(0, 10, size=n)
data2 = stats.norm.rvs(5, 1.5, size=n)
data3 = np.r_[stats.norm.rvs(2, 0.25, size=int(n/2)), stats.norm.rvs(8, 0.5, size=int(n/2))]
data4 = stats.norm.rvs(6, 0.5, size=n)

fig = make_subplots(rows=1, cols=4)

fig.add_trace(go.Histogram(x=data1, name='A', nbinsx=30, marker=dict(line=dict(color='black', width=1))), row=1, col=1)
fig.add_trace(go.Histogram(x=data2, name='B', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=2)
fig.add_trace(go.Histogram(x=data3, name='C', nbinsx=45, marker=dict(line=dict(color='black', width=1))), row=1, col=3)
fig.add_trace(go.Histogram(x=data4, name='D', nbinsx=15, marker=dict(line=dict(color='black', width=1))), row=1, col=4)

fig.update_layout(height=300, width=750, title_text="Row of Histograms")
fig.update_xaxes(title_text="A", row=1, col=1)
fig.update_xaxes(title_text="B", row=1, col=2)
fig.update_xaxes(title_text="C", row=1, col=3)
fig.update_xaxes(title_text="D", row=1, col=4)
fig.update_xaxes(range=[-0.5, 10.5])

for trace in fig.data:
    trace.xbins = dict(start=0, end=10)
    
# This code was produced by just making requests to Microsoft Copilot
# https://github.com/pointOfive/stat130chat130/blob/main/CHATLOG/wk3/COP/SLS/0001_concise_makeAplotV1.md

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

````python
# Code used to find their respective mean and variance

for i, x in enumerate([data1, data2, data3, data4],1):
    print("Mean " + str(i) + ": " + str(np.mean(x)))
    
print('\n')
    
for i, x in enumerate([data1, data2, data3, data4],1):
    print("Variance " + str(i) + ": " + str(np.var(x, ddof=1)))
````

For a right skewness, the mean is less than the median.
For a left skewness, the mean is larger than the median.



This is because for a right skewed distribution, there are extreme values on the right (very large outliers!) affecting the value of mean significantly. However, the outliers only insignificantly affect the value of median as the position of the median is independent of the data values. The below code (given code) executed generates the corresponding graph.

````python
# By executing the given code, we get a right-skewed histogram.

from scipy import stats
import pandas as pd
import numpy as np
  
sample1 = stats.gamma(a=2,scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data")
fig1.show(renderer="png")

sample1.mean()
np.quantile(sample1, [0.5]) # median

````
